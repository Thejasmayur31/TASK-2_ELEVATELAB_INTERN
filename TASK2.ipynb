# TASK 2:Exploratory Data Analysis (EDA)
 DATASET : NETFLIX AND TV SHOWS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info() # getting basic info
```

```python
df.describe(include='all')
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
df['duration'].fillna(df['duration'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Movies vs TV Shows')
plt.show()
```

```python
sns.histplot(data=df, x='release_year', bins=30, kde=True) # visualizing using histplot which is similer to histogram that is related to time / dates
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.show()
```

```python
df['duration_time'] = df['duration'].str.extract('(\d+)').astype(float) # extract('(\d+)')  : converting to the regular expression i.e time in the dataset has 2 types  1 is movie with min and tv shows with seasons
Movies = df[df['type'] == 'Movie'] # considering only movies
sns.boxplot( data=Movies, x='duration_time')
plt.title('Box Plot of Movie Durations')
plt.xlabel('Duration (minutes)')
```

```python
top_country = df['country'].value_counts().head(10) # counts how many times each country produces the shows
plt.figure(figsize=(10, 6))
sns.barplot(x=top_country.index, y=top_country.values)
plt.title('Top 10 Countries with the Most Content')
plt.xlabel('Country')
plt.show()
```

```python
corr = df[['release_year','duration_time']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation between Release Year and Duration')
plt.show()
```
